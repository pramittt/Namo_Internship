```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Bihar_Election_Results.csv")
```

```python
data=df.values.tolist()
n=len(data)
m=len(data[0])
```

```python
p_list=list(set(df['Party'].values.tolist()))
y_list=list(set(df['Election Year'].values.tolist()))
sn_list=list(set(df['AC No'].values.tolist()))
y_list.sort()
```

# Solution 1:

```python
res=[[0 for i in range(len(y_list))] for j in range(len(p_list))]

for i in range(n):
    if(data[i][df.columns.get_loc("Party")]!='Independent'):
        res[p_list.index(data[i][df.columns.get_loc("Party")])][y_list.index(data[i][df.columns.get_loc("Election Year")])]+=int(data[i][df.columns.get_loc("Votes")])

res=pd.DataFrame(res)
res.columns=y_list
res.insert(loc=0, column='Party', value=p_list)

l=[]
for x in y_list:
    l_tmp=[]
    tmp=res[['Party',x]]
    tmp=tmp.sort_values(by=[x],ascending=False)
    tmp=tmp.reset_index(drop=True)
    l_tmp.append(x)
    l_tmp.append(tmp['Party'][0])
    l_tmp.append(tmp[x][0])
    l.append(l_tmp)
res1=pd.DataFrame(l)
res1
```

# Solution 4:

```python
tmp=res3.loc[res3[0]=='2015']
tmp=tmp.reset_index(drop=True)

top8=list(tmp[1].values.tolist())

res4=[[100000000 for i in range(len(top8))] for j in range(len(sn_list))]

for i in range(n):
    if((data[i][df.columns.get_loc("Party")] in top8)==True):
        if(res4[data[i][df.columns.get_loc("AC No")]-1][top8.index(data[i][df.columns.get_loc("Party")])]>data[i][df.columns.get_loc("Votes")]):
            res4[data[i][df.columns.get_loc("AC No")]-1][top8.index(data[i][df.columns.get_loc("Party")])]=data[i][df.columns.get_loc("Votes")]

for i in range(len(top8)):
    for j in range(len(sn_list)):
        if(res4[j][i]==100000000):
            res4[j][i]='NaN'
            
res4=pd.DataFrame(res4)
res4.columns=top8

sn=[""]*len(sn_list)

for i in range(n):
    sn[data[i][df.columns.get_loc("AC No")]-1]=data[i][df.columns.get_loc("AC name")]
    
res4.insert(loc=0, column='Seat name', value=sn)

res4
```

# Solution 5:

```python
res5=[[0 for i in range(1)] for j in range(len(sn_list))]


for i in range(n):
    if(data[i][df.columns.get_loc("Position")]==1):
        if(float(data[i+1][df.columns.get_loc("Votes %")].strip('%'))>=10 and float(data[i+2][df.columns.get_loc("Votes %")].strip('%'))>=10):
            res5[data[i][df.columns.get_loc("AC No")]-1][0]+=1
            
res5=pd.DataFrame(res5)
res5.insert(loc=0, column='Seat name', value=sn)
res5=res5.loc[res5[0]==4]
res5=res5.drop(columns=[0])
res5=res5.reset_index(drop=True)
res5
```

# Solution 6:

```python
res6=[[1 for i in range(1)] for j in range(len(sn_list))]

for i in range(n):
    if(data[i][df.columns.get_loc('Position')]==1):
        if(len(res6[data[i][df.columns.get_loc('AC No')]-1])==1):
            res6[data[i][df.columns.get_loc('AC No')]-1].append(data[i][df.columns.get_loc('Party')])
        else:
            if(res6[data[i][df.columns.get_loc('AC No')]-1][1]==data[i][df.columns.get_loc('Party')]):
                res6[data[i][df.columns.get_loc('AC No')]-1][0]+=1
                
res6=pd.DataFrame(res6)
res6.insert(loc=0, column='Seat name', value=sn)
res6=res6.loc[res6[0]==4]
res6=res6.drop(columns=[0])
res6.columns=['Seat name','Party']
res6=res6.reset_index(drop=True)
res6
```
